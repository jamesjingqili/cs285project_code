```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_returns = np.load("./all_returns.npy")
```

```python
fig = plt.figure()

for i in range(all_returns.shape[0]):
    plt.plot(np.arange(all_returns.shape[1]), all_returns[i,:])

plt.legend(['task '+str(i+1) for i in range(all_returns.shape[0])])
plt.ylabel('Reward')
plt.xlabel('Training iteration')
plt.show()
```
